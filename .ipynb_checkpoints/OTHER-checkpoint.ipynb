# Preprocessing for Columns,which were done explained:
    PUlocationID & DOlocationID: Normally business/office/market areas will be busy than normal area. SO I grouped them into 3 categories, based on 
                                 their busy status, which is value counts() and categorical variables 'DOarea', and 'PUarea' were created. 
                                 
    trip_distance:               When trip distancs is more than certain km, they charge extra. But that extra depends on categories. SO i created 3 
                                 categories based  on trip_distance. trip_distance is important parameter.  4 outliers of more than 1000 km removed. 
                                 
    tpep_pickup_datetime  & tpep_dropoff_datetime:
                                 Pickup and Drop timings are important like trip distance. Hence, they were grouped into 4 categories first. Then 
                                 weekdays or weekends influence rate. Also months or particular dates also influence. Hence all details are extracted. 
                                 Then time objects are deleted.
                                 
    VendorID and RateCodeID:     They are IDs fo vendor and rate card.  Rate code Id has missing values. Hence to be imputed using mean strategy.  
    
    store_and_fwd_flag & payment_type  : They are categorical variables, without any ranking among them. Hence categorised using one hot method. They                                  have unknown and missing values. That have to be filled up suitably.
    
    extra, tolls_amount,
    tips_amount
    improvement_surcharge,
    congestion_surcharge
    airport fee:                 They are numerical, and essential to keep.congestion_surcharge and airport fee has null values, to be imputed.
    
    tip_amount                  : Numerical type. 1 outlier of 464 which is equivalent ot fare is removed.
    
    passenger_count             : Numerical type. Though 9 is there which is unrealistic. 
    
    total_amount                : It has lot of negative values. 
    
    

# Preprocessing for passenger_count:
     All 0 values (2828 nos.) are replaced with predominant 1.
    All high values 9,8 are replaced with 1.
    All null values are replaced with 1.
    predominant is 1.

```python
print(train_data['passenger_count'].value_counts())
train_data.info()
train_data.loc[train_data['passenger_count'] == 9, 'passenger_count' ] = train_data['passenger_count'].mean()
train_data.loc[train_data['passenger_count'] == 8, 'passenger_count' ] = train_data['passenger_count'].mean()
train_data.loc[train_data['passenger_count'] == 0, 'passenger_count' ] = train_data['passenger_count'].mean()
print(train_data['passenger_count'].isnull().sum())
train_data.loc[train_data['passenger_count'].isna(), 'passenger_count'] = train_data['passenger_count'].mean()
print(train_data['passenger_count'].isnull().sum())
```

# Preprocessing for PULocationID and DOLocationID:
     3 bins 'PUALow',PUAMedium','PUAHeavy' are created from taking values from describe()
     At first value counts for each area was taken to find out their distribution.
     Then created separate column using mapping.
     Finally PUarea bins were created using apply method. 
     Bins are decided using it has mean of 663.86, 25% is 646, 75% is 680. So decided to  use them for bins.     
     Histogram shows the distribution.
    


za = train_data['PULocationID'].value_counts()
train_data['puvc']=train_data['PULocationID'].map(za)
train_data['puvc'].head()
print(train_data['puvc'].describe())
def puvc_to_grade(x):
    if x <646.1:
        return 'PUALow'
    elif x >646 and x < 680.1:
        return 'PUAMedium'
    else:
        return 'PUAHeavy'
train_data['PUarea'] = train_data['puvc'].apply(puvc_to_grade)
train_data['PUarea'].head()
train_data['PUarea'].hist(bins = 5,ec='red',color='yellow',grid=True)
